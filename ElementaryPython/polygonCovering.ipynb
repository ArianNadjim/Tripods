```python
"""Polygon covering problem exercise
Suppose we are given the vertices of a convex polygon (in order) and a radius r.
We want to find locations of balls of radius r that cover the polygon."""

import matplotlib.pyplot as plt
from numpy import random


# Define the vertices counter-clockwise and the radius of the balls.
vertices = [[1,0], [0,1], [-1,0], [0,-1]]
ballRadius = .5


# Populate x- and y- value arrays.
xValues = []
yValues = []
for v in vertices:
    plt.plot(v[0], v[1], 'o', color='black')
    xValues.append(v[0])
    yValues.append(v[1])

# Draw the shape.
plt.plot(xValues, yValues, color = 'black')
plt.plot([xValues[len(xValues)-1], xValues[0]],[yValues[len(yValues)-1], yValues[0]], color='black')

# Pick some random points within our shape
# Gives each vertex a weight of (random float between 0 and 1)/(number of vertices)
# Maybe some clever person can find a better way to pick the centers of our balls?
centers = []
circles = []
for i in range(4):
    centers.append([0,0])
    for v in vertices:
        rando = random.rand()/len(vertices)
        centers[i][0] = centers[i][0] + rando*v[0]
        centers[i][1] = centers[i][1] + rando*v[1]
    print(centers[i])
    tempCircle = plt.Circle((centers[i][0],centers[i][1]), radius = ballRadius, fill=False)
    circles.append(tempCircle)
    
# Draw the covering balls
fig = plt.gcf()
ax = fig.gca()
for c in circles:
    ax.add_artist(c)
```
