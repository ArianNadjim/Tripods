```python
# PDF

import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm

x=np.linspace(-6,6,1000)

# Generates 1000 equally spaced numbers from -6 to 6

fx=norm.pdf(x, loc=0, scale=1)

# Normal distrubution probability density function with mean 0 and standard deviation 1. 

plt.plot(x,fx)

# 
```

```python
# CDF 

x=np.linspace(-6,6,1000)

# Generates 1000 equally spaced numbers from -6 to 6

Fx=norm.cdf(x, loc=0, scale=1)

# Normal distribution cumulative distribution function 

plt.plot(x,Fx)

```

```python
# LOGPDF

import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm

x=np.linspace(-6,6,1000)

# Generates 1000 equally spaced numbers from -6 to 6

logfx=norm.logpdf(x, loc=0, scale=1)

# Normal distribution logarithm of the probability distribution function 

plt.plot(x,logfx)

```

```python
# CONVOLUTION 

from PIL import Image

im=Image.open("lena.jpg")

gray=np.mean(im,axis=2)

x=np.linspace(-6,6,50)

fx=norm.pdf(x, loc=0, scale=1)

filt=np.outer(fx,fx)

plt.imshow(filt,cmap="gray")
```

```python
from scipy.signal import convolve2d

out=convolve2d(gray,filt)

plt.subplot(1,2,1)

plt.imshow(gray,cmap="gray")

plt.subplot(1,2,2)

plt.imshow(out,cmap="gray")
```
