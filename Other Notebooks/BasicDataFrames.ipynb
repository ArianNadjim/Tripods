```python
# BUILDING A DATAFRAME BY HAND

import pandas as pd

data1 = [['Alex', 52, "Male", 6], ['Josh', 17, "Male", 6.25], ['Nathan', 15, "Male", 6.1], 
         ['Shannon', 38, "Female", 5.8], ['Simon', 6, "Male",4]]

# Creates a list needed for the creationg of the dataframe
  
df1 = pd.DataFrame(data1, columns = ['Name', 'Age', "Sex", "Height"])

# Creates the dataframe
  
df1

# Prints the dataframe 

#ANOTHER METHOD

data2 = {'Name':["Alex","Shannon","Josh","Nathan","Simon"], 'Age':[52,38,17,15,6]} 

# Another way of generating the same data 
  
df2 = pd.DataFrame(data2) 

# Creates the dataframe. 
  
df2 

# Displays the dataframe 

df1.set_index('Name')

# Sets "Name" as the index column 

```

```python
import pandas as pd

# Creatting several dictionaries that will be turned into data frames

data1 = {'Name':["Alex","Shannon"], 
         'Age':[52,38]} 

data2 = {'Name':["Josh","Nathan","Simon"], 
         'Age':[17,15,6]} 

data3 = {'Name':["Misha","Sveta"],'Age':[83,75]}

data4 = {'Name':["Steve","Debbie"],'Age':[69,69]}


# Creating the corresponding dataframes 
  
df1 = pd.DataFrame(data1) 

df2 = pd.DataFrame(data2)

df3 = pd.DataFrame(data3)

df4 = pd.DataFrame(data4)

# Creating a list of these dataframes 

D=[df1,df2,df3,df4]

# Creating a dataframe that merges the dataframes above, sorts the values by age, and sets "Name" to be the index. 
# Note that python methods can be piled on top of one another. Note that if we set ascending=True, we would obtain 
# sorting by age from youngest to oldest. What is the default? Check! 

df1234=pd.concat(objs=D,ignore_index=True).sort_values(by='Age',ascending=False).set_index('Name')

df1234
```
