```python
# LINE PLOT

import numpy as np 
import matplotlib.pyplot as plt

x=np.linspace(0,20,1000)

# This creates a one-dimensional array with 1000 equally spaced points between 0 and 20. 

y=np.sin(x)+0.2*x 

# We will be plotting this in a moment, so we generated some data for the vertical coordinate. 

plt.plot(x,y)

# This plots the data above. 

plt.xlabel("Input")
plt.ylabel("Output")

# This labels the axes in your plot. 

plt.title("My plot");

# This labels the whole plot. The semi-colon prevents any extra junk from being printed before the plot. Try it with
#$ and without the semi-colon. 
```

```python
# SCATTER PLOT

import numpy as np 
import matplotlib.pyplot as plt

X=np.random.randn(100,2)

# Recall that this generates 100 2-vectors with random entries in accordance with the normal distribution. 

plt.scatter(X[:,0],X[:,1])

# This creates a scatter plot of X. Note that X[:,0] denotes the first coordinates from X and X[:,1] denotes the 
# second coordinates from X. 

```

```python
# SCATTER PLOT- more complicated operations 

import numpy as np 
import matplotlib.pyplot as plt

X=np.random.randn(200,2)

# Generates 200 random 2-vectors in accordance with the normal distribution. 

X[:50] +=3

# Takes all the rows from index 0 to index 50 and adds 3 to all the elements. 

Y=np.zeros(200)

# Generates 200 zeroes. 

Y[:50]=1

# Takes all the rows from index 0 to index 50 and assigns a label 1 to them. 

plt.scatter(X[:,0],X[:,1], c=Y)



```

```python
# HISTOGRAMS 

X=np.random.randn(10000)

plt.hist(X,bins=50);

# Generates a histogram of 10000 points chosen in accordance with the normal distribution. The bins=50 increases the 
# number of bins. Play with it. 

Y=np.random.random(10000)

plt.hist(Y,bins=50);

# Generates a histogram of 10000 points chosen in accordance with the uniform distribution. The plot below shows 
# both, with the normally distributed points in blue, and the uniformly distributed points in orange. 
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def prime_count(n):
    for N in range(1,n+1):
        prime = {x for x in range(2, N) if all(x % y != 0 for y in range(2,int(math.sqrt(x))+1))}
    return len(prime)


for k in range(2,2000):
    print(prime_count(k)*np.log(k)/k)


    




    
    

    
    



```
